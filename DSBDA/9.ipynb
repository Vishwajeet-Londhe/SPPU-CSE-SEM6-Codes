```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from seaborn import load_dataset
```

```python
#titanic dataset 
dataset = sns.load_dataset('titanic')
```

```python
#tips dataset 
tips = load_dataset("tips")
```

```python
dataset.head()
```

```python
sns.countplot(x='survived', data=dataset)
plt.show()
```

```python
dataset['sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(dataset['age'], bins=5) 
plt.show()
```

```python
sns.distplot(dataset['age'])  
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

```python
sns.barplot(x='pclass', y='age', hue='sex', data=dataset)
plt.title("AverageAge by Class and Gender")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
plt.title("Age Distribution by Sex")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=dataset)
plt.title("Age distribution by Sex and Survival")
plt.show()
```

```python
sns.distplot(dataset[dataset['survived'] == 0]['age'], hist=False, color="blue")
sns.distplot(dataset[dataset['survived'] == 1]['age'], hist=False, color="orange") 
plt.show()
```

```python
pd.crosstab(dataset['pclass'], dataset['survived'])
```

```python
sns.heatmap(pd.crosstab(dataset['pclass'], dataset['survived']))
```

```python
sns.clustermap(pd.crosstab(dataset['parch'], dataset['survived']))
plt.show()
```
